```python
# import all necessary modules
import pandas as pd

import string
import numpy as np

import webvtt
import textwrap as tp
```

# Import a text file

```python
file_path = "captions_text.vtt"

captions = webvtt.read(file_path)


captions_dict = {}

for caption in captions:
    start_time = caption.start
    text = caption.text
    captions_dict[start_time] = text

```

# Converting the data stored in dictionary into a normal paragraph

```python
values = captions_dict.values()

string_data = ""
for i in values:
    string_data += i + " "


wrapped_string = tp.fill(string_data, width=80)
print(wrapped_string)

print(type(string_data))
```

# Replace 4-letter words and save to disk

```python
# convert newtext to words
words = wrapped_string.split()

for i in range(len(words)):
    if len(words[i])==4 and words[i].isalpha():
        words[i] = words[i][0] + "*" * 3 
        
        
# putting back together from separate words into one string

cleaned_words = " ".join(words)
print(cleaned_words)
```

```python
# Save back to disk
with open("clean_text.txt","w") as wf:
  wf.write(cleaned_words)
```

# Bonus: Readability of scrambled words

```python
# A sentence
text = "I have often wondered whether it's possible to read an entire book faster if the useless letters were removed."

```

```python
# split the text into words
words = text.split()
print(words)
```

```python
# randomly scramble all letters within each word
out =[]
for i in range(len(words)):
    
    this_word = words[i]
    rdrng = np.random.permutation(len(this_word))
    check = "".join([this_word[i] for i in rdrng])
    out.append(check)
print(out)
```

```python
# put back together into one string
one_str = " ".join(out)
print(one_str)
```

```python
# scramble all letters except first and last (only words with >3 letters)

# recover original
wordsfl = text.split()

# try again
for wi in range(len(wordsfl)):
  
  # skip if too short
  if len(wordsfl[wi])<4:
    continue

  # extract subword
  thisword = wordsfl[wi][1:-1]
  idx = np.random.permutation(len(thisword))
  wordsfl[wi] = wordsfl[wi][0] + ''.join([thisword[i] for i in idx]) + wordsfl[wi][-1]

' '.join(words)
```

```python
# replace with random letters

# all letters
allletters = string.ascii_lowercase

# recover original
words = text.split()

# try again
for wi in range(len(words)):
  
  # skip if too short
  if len(words[wi])<4:
    continue

  # extract subword
  thisword = words[wi][1:-1]
  idx = np.random.randint(0,high=len(allletters),size=len(thisword))
  words[wi] = words[wi][0] + ''.join([allletters[i] for i in idx]) + words[wi][-1]

' '.join(words)
```
